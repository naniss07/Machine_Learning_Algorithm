```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from time import sleep
from random import randint

# ChromeDriver'ı başlatmak için ayarları yapılandır
chrome_options = Options()
chrome_options.add_argument("--headless")  # Tarayıcıyı başlatırken görünür yapmamak için
chrome_options.add_argument("--disable-gpu")  # GPU kullanımını devre dışı bırak
chrome_options.add_argument("--no-sandbox")  # Linux sistemlerinde sandbox'ı devre dışı bırak

# Chrome Driver servisini başlat
driver_path = "C:\chromedriver.exe"
service = Service(driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Liste başlangıçları
post_links = []
prices = []
addresses = []
bedrooms = []
bathrooms = []
sqft = []

# Sayfa aralığı
pages = range(1, 135)  # 1. sayfadan başla

for page in pages:
    print(f"\nSayfa {page} için veriler çekiliyor...")
    url = f"https://www.zoocasa.com/vancouver-bc-real-estate?page={page}"
    driver.get(url)

    # Dinamik yükleme bekleme (kartlar görünene kadar)
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "style_component__dHZly"))
        )
    except Exception as e:
        print(f"Sayfa {page} yüklenirken hata: {e}")
        continue

    # Listing kartlarını bul
    listing_cards = driver.find_elements(By.CLASS_NAME, "style_component__dHZly")
    print(f"Sayfa {page}: {len(listing_cards)} kart bulundu.")

    for card in listing_cards:
        # Linki al
        try:
            link = card.find_element(By.TAG_NAME, "a")
            post_link = link.get_attribute("href")
            post_links.append(post_link if post_link else 'Link bulunamadı')
        except:
            post_links.append('Link bulunamadı')

        # Fiyatı al
        try:
            price = card.find_element(By.XPATH, ".//div[@data-testid='listingPrice']")
            prices.append(price.text.strip() if price.text else 'Fiyat bulunamadı')
        except:
            prices.append('Fiyat bulunamadı')

        # Adresi al
        try:
            address = card.find_element(By.CLASS_NAME, "style_street-address__m_EWt")
            addresses.append(address.text.strip() if address.text else 'Adres bulunamadı')
        except:
            addresses.append('Adres bulunamadı')

        # Yatak odası sayısını al
        try:
            bed = card.find_element(By.XPATH, ".//*[@data-testid='listingBeds']")
            bedrooms.append(bed.text.strip() if bed.text else 'Yatak odası bilgisi bulunamadı')
        except:
            bedrooms.append('Yatak odası bilgisi bulunamadı')

        # Banyo sayısını al
        try:
            bath = card.find_element(By.XPATH, ".//*[@data-testid='listingBathrooms']")
            bathrooms.append(bath.text.strip() if bath.text else 'Banyo bilgisi bulunamadı')
        except:
            bathrooms.append('Banyo bilgisi bulunamadı')

        # Alan bilgisini al
        try:
            sqft_area = card.find_element(By.XPATH, ".//*[@data-testid='listingSqft']")
            sqft.append(sqft_area.text.strip() if sqft_area.text else 'Alan bilgisi bulunamadı')
        except:
            sqft.append('Alan bilgisi bulunamadı')


    # Rastgele bekleme
    sleep(randint(1, 3))
    print(f"Sayfa {page} başarıyla işlendi!")

# Tarayıcıyı kapat
driver.quit()

# Konsola verileri yazdır
print("\nKonsola Yazdırılan Veriler:")
for i in range(len(post_links)):
    print(f"Post Link: {post_links[i]} , Price: {prices[i]}, Address: {addresses[i]}, Bedroom Count: {bedrooms[i]}, Bathroom Count: {bathrooms[i]}, Sqft: {sqft[i]}")

# Veriyi pandas DataFrame'e dönüştür
df = pd.DataFrame({
    'Post Link': post_links,
    'Price': prices,
    'Address': addresses,
    'Bedroom Count': bedrooms,
    'Bathroom Count': bathrooms,
    'Sqft': sqft
})

# Excel dosyasına yaz
df.to_excel('real_estate_data_selenium.xlsx', index=False)

print("\nVeriler başarıyla Excel dosyasına yazdırıldı!")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
from time import sleep
from random import randint

# ChromeDriver'ı başlatmak için ayarları yapılandır
chrome_options = Options()
chrome_options.add_argument("--headless")  # Tarayıcıyı başlatırken görünür yapmamak için
chrome_options.add_argument("--disable-gpu")  # GPU kullanımını devre dışı bırak
chrome_options.add_argument("--no-sandbox")  # Linux sistemlerinde sandbox'ı devre dışı bırak

# Chrome Driver servisini başlat
driver_path = "C:\\chromedriver.exe"  # Windows için yol
service = Service(driver_path)
driver = webdriver.Chrome(service=service, options=chrome_options)

# Liste başlangıçları
post_links = []

# Ana sayfadan ilan URL'lerini çek
pages = range(1, 132)  # Sayfa aralığı (1'den 4'e kadar örnek için)

for page in pages:
    print(f"\nSayfa {page} için veriler çekiliyor...")
    url = f"https://www.zoocasa.com/vancouver-bc-real-estate?page={page}"
    driver.get(url)

    # Dinamik yükleme bekleme (kartlar görünene kadar)
    try:
        WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.CLASS_NAME, "style_component__dHZly"))
        )
    except Exception as e:
        print(f"Sayfa {page} yüklenirken hata: {e}")
        continue

    # Listing kartlarını bul
    listing_cards = driver.find_elements(By.CLASS_NAME, "style_component__dHZly")
    print(f"Sayfa {page}: {len(listing_cards)} kart bulundu.")

    for card in listing_cards:
        # Linki al
        try:
            link = card.find_element(By.TAG_NAME, "a")
            post_link = link.get_attribute("href")
            post_links.append(post_link if post_link else 'Link bulunamadı')
        except:
            post_links.append('Link bulunamadı')

    # Rastgele bekleme
    sleep(randint(1, 3))
    print(f"Sayfa {page} başarıyla işlendi!")

# İlan detaylarından veri çekme
detailed_prices = []
detailed_addresses = []
detailed_bedrooms = []
detailed_bathrooms = []
detailed_sqft = []
detailed_garages = []
detailed_ages = []

for link in post_links:
    try:
        # İlan detay sayfasına git
        driver.get(link)
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.TAG_NAME, "body"))
        )
        print(f"{link} sayfasından veri çekiliyor...")

        # Fiyat bilgisi
        try:
            price = driver.find_element(By.XPATH, ".//div[@data-testid='listingPriceModal']")
            detailed_prices.append(price.text.strip())
        except:
            detailed_prices.append('Fiyat bulunamadı')

        # Adres bilgisi
        try:
            address = driver.find_element(By.XPATH, ".//h1[@data-testid='streetAddress']")
            detailed_addresses.append(address.text.strip())
        except:
            detailed_addresses.append('Adres bulunamadı')

        # Yatak odası bilgisi
        try:
            bed = driver.find_element(By.XPATH, ".//span[@data-testid='listingBedIcon']")
            detailed_bedrooms.append(bed.text.strip())
        except:
            detailed_bedrooms.append('Yatak odası bilgisi bulunamadı')

        # Banyo bilgisi
        try:
            bath = driver.find_element(By.XPATH, ".//span[@data-testid='listingBathIcon']")
            detailed_bathrooms.append(bath.text.strip())
        except:
            detailed_bathrooms.append('Banyo bilgisi bulunamadı')

        # Alan bilgisi
        try:
            sqft_area = driver.find_element(By.XPATH, ".//span[@data-testid='listingDimensionsIcon']")
            detailed_sqft.append(sqft_area.text.strip())
        except:
            detailed_sqft.append('Alan bilgisi bulunamadı')

        # Garaj bilgisi
        try:
            garage = driver.find_element(By.XPATH, ".//span[@data-testid='listingCarIcon']")
            detailed_garages.append(garage.text.strip())
        except:
            detailed_garages.append('Garaj bilgisi bulunamadı')

        try:
            age_element = driver.find_element(By.CSS_SELECTOR, "section > div:nth-child(8) > span:nth-child(2)")
            detailed_ages.append(age_element.text.strip())
        except:
            detailed_ages.append('Yıl bilgisi bulunamadı')


        print(f"{link} başarıyla işlendi.")
        sleep(randint(1, 2))  # Rastgele bekleme

    except Exception as e:
        print(f"{link} işlenirken hata oluştu: {e}")
        # Hata durumunda listelere "bulunamadı" ekle
        detailed_prices.append('Hata oluştu')
        detailed_addresses.append('Hata oluştu')
        detailed_bedrooms.append('Hata oluştu')
        detailed_bathrooms.append('Hata oluştu')
        detailed_sqft.append('Hata oluştu')


# Tarayıcıyı kapat
driver.quit()

# Veriyi pandas DataFrame'e dönüştür
df = pd.DataFrame({
    'Post Link': post_links,
    'Detailed Price': detailed_prices,
    'Detailed Address': detailed_addresses,
    'Detailed Bedroom Count': detailed_bedrooms,
    'Detailed Bathroom Count': detailed_bathrooms,
    'Detailed Sqft': detailed_sqft,
    'Detailed Garage Count' : detailed_garages,
    'Detailed Ages' : detailed_ages
})


```
