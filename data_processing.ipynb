```python
import pandas as pd
import numpy as np

file_path = "updated_file.xlsx"
df = pd.read_excel(file_path)
df.head(3)
```

```python
df_cleaned = df.dropna()
df = df_cleaned
df.head(3)
```

```python
df['Detailed Price'] = df['Detailed Price'].str.replace('$', '', regex=False)
df['Detailed Price'] = df['Detailed Price'].str.replace(',', '', regex=False)
```

```python
df.head(2)
```

```python
new_order = ['Detailed Bedroom Count', 'Detailed Bathroom Count', 'Detailed Garage Count' , 'Detailed Ages', 'Detailed Sqft','Latitude', 'Longitude', 'Detailed Price', 'Detailed Address', 'Post Link']
df = df[new_order]
df.head(2)
```

```python
df['Detailed Bedroom Count'] = df['Detailed Bedroom Count'].astype(float)
df['Detailed Bathroom Count'] = df['Detailed Bathroom Count'].astype(float)
df['Detailed Garage Count'] = df['Detailed Garage Count'].astype(float)
df['Detailed Ages'] = df['Detailed Ages'].astype(float)

df.head(3)
```

```python
df['Detailed Sqft'] = df['Detailed Sqft'].str.replace(' sqft', '', regex=True)
df['Detailed Sqft'] = df['Detailed Sqft'].astype(float)
df.head(3)
```

```python
df['Detailed Sqft'] = (df['Detailed Sqft'] * 0.092903).round()
df['Detailed Ages'] = 2025 - (df['Detailed Ages'])
df.head(3)
```

```python
df2 = df[['Detailed Bedroom Count', 'Detailed Bathroom Count', 'Detailed Ages', 'Detailed Garage Count','Detailed Sqft', 'Latitude', 'Longitude', 'Detailed Price']]
df2['Detailed Price'] = df2['Detailed Price'].astype(float)
df2['Detailed Sqft'] = df2['Detailed Sqft'].astype(float)

df2 = df2.rename(columns={'Detailed Bedroom Count': 'Bedroom_count', 'Detailed Bathroom Count': 'Bathroom_count', 'Detailed Garage Count': 'Garage_count', 'Detailed Ages': 'Ages', 'Detailed Sqft':'Sqm', 'Detailed Price': 'Price'})

df2.head(2)
```

```python
df2.to_excel("cleaned_data_excel.xlsx", index=False)
```
