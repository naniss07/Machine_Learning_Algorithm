```python
import pandas as pd
import numpy as np
dataset = pd.read_excel("cleaned_data_excel.xlsx")
dataset.head(2)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import (
    Ridge,
    ElasticNet,
    Lasso,
    BayesianRidge,
    HuberRegressor,
    LinearRegression
)
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import QuantileRegressor

# Veriyi hazırlama
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
y = y.reshape(-1, 1)

# Veriyi eğitim ve test setlerine ayırma
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)

# Ortak metrik hesaplama fonksiyonu
def evaluate_model(y_test, y_pred, model_name):
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    # Çizgi grafiklerini çizdirme
    plt.figure(figsize=(10, 6))
    plt.plot(y_test, label='Gerçek Değerler', color='blue')
    plt.plot(y_pred, label='Tahmin Edilen Değerler', color='red', linestyle='--')
    plt.title(f'{model_name} - Gerçek vs Tahmin Değerleri')
    plt.xlabel('Test Verisi')
    plt.legend()
    plt.show()
    
    print(f"{model_name} - Mean Absolute Error (MAE): {mae}")
    print(f"{model_name} - Mean Squared Error (MSE): {mse}")
    print(f"{model_name} - R-squared (R²): {r2}")
    print("-" * 50)


# Decision Tree Regression
dt_regressor = DecisionTreeRegressor(random_state=5)
dt_regressor.fit(X_train, y_train)
y_pred = dt_regressor.predict(X_test)
evaluate_model(y_test, y_pred, "Decision Tree Regression")

# Random Forest Regression
rf_regressor = RandomForestRegressor(n_estimators=7, random_state=5)
rf_regressor.fit(X_train, y_train)
y_pred = rf_regressor.predict(X_test)
evaluate_model(y_test, y_pred, "Random Forest Regression")

# Linear Regression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)
y_pred = linear_regressor.predict(X_test)
evaluate_model(y_test, y_pred, "Linear Regression")

# KNN Regression
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_model.fit(X_train, y_train)
y_pred = knn_model.predict(X_test)
evaluate_model(y_test, y_pred, "KNN Regression")

# Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(X_train, y_train)
y_pred = ridge_model.predict(X_test)
evaluate_model(y_test, y_pred, "Ridge Regression")

# Elastic Net Regression
elastic_net_model = ElasticNet(alpha=1.0, l1_ratio=0.5)
elastic_net_model.fit(X_train, y_train)
y_pred = elastic_net_model.predict(X_test)
evaluate_model(y_test, y_pred, "Elastic Net Regression")

# Huber Regression
huber_reg = HuberRegressor(max_iter=1000, epsilon=1.35)
huber_reg.fit(X_train, y_train)
y_pred = huber_reg.predict(X_test)
evaluate_model(y_test, y_pred, "Huber Regression")

# Lasso Regression
lasso_model = Lasso(alpha=0.5)
lasso_model.fit(X_train, y_train)
y_pred = lasso_model.predict(X_test)
evaluate_model(y_test, y_pred, "Lasso Regression")

# Bayesian Ridge Regression
bayesian_ridge = BayesianRidge()
bayesian_ridge.fit(X_train, y_train)
y_pred = bayesian_ridge.predict(X_test)
evaluate_model(y_test, y_pred, "Bayesian Ridge Regression")

# Quantile Regression
quantile_reg = QuantileRegressor(quantile=0.5, alpha=0)
quantile_reg.fit(X_train, y_train)
y_pred = quantile_reg.predict(X_test)
evaluate_model(y_test, y_pred, "Quantile Regression")
```
