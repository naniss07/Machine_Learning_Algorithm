```python
import pandas as pd
import numpy as np

file_path = "updated_file.xlsx"
df = pd.read_excel(file_path)
df.head(3)
```

```python
df_cleaned = df.dropna()
df = df_cleaned
df.head(3)
```

```python
df['Detailed Price'] = df['Detailed Price'].str.replace('$', '', regex=False)
df['Detailed Price'] = df['Detailed Price'].str.replace(',', '', regex=False)
```

```python
df.head(2)
```

```python
new_order = ['Detailed Bedroom Count', 'Detailed Bathroom Count', 'Detailed Garage Count' , 'Detailed Ages', 'Detailed Sqft','Latitude', 'Longitude', 'Detailed Price', 'Detailed Address', 'Post Link']
df = df[new_order]
df.head(2)
```

```python
df['Detailed Bedroom Count'] = df['Detailed Bedroom Count'].astype(float)
df['Detailed Bathroom Count'] = df['Detailed Bathroom Count'].astype(float)
df['Detailed Garage Count'] = df['Detailed Garage Count'].astype(float)
df['Detailed Bathroom Count'] = df['Detailed Bathroom Count'].astype(float)

df
```

```python
def calculate_average(sqft_value):
    sqft_value = sqft_value.replace(' sqft', '')
    
    sqft_value = sqft_value.replace('–', '-')
    
    if '-' in sqft_value:
        start, end = sqft_value.split('-')
        return (int(start) + int(end)) / 2
    return int(sqft_value)

df['Sqft'] = df['Sqft'].apply(calculate_average)
df
```

```python
df2 = df[['Bedroom Count', 'Bathroom Count', 'Sqft', 'Latitude', 'Longitude', 'Price']]
df2['Price'] = df2['Price'].astype(float)

df2 = df2.rename(columns={'Bedroom Count': 'Bedroom_count', 'Bathroom Count': 'Bathroom_count'})

df2.head(2)
```

```python

df2.to_csv("cleaned_data", index=False)
```
